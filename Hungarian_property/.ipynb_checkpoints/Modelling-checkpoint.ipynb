```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
import datetime
```

```python
db = pd.read_csv('VkerDB.csv')
db['source'] = db.source.astype('category')
db.sort_values('date', inplace = True)
db.reset_index(drop = True, inplace = True)
db.shape
```

```python
db.info()
```

```python
#price prediction for give date
```

```python
last = db.loc[db['date'] == db['date'].unique()[-1], ['price', 'room','sqm']]
print('The available advetisments for the prediction: ', last.shape[0])
#last = last.loc[(last['price'] < last['price'].quantile(0.95))]
print('')
```

```python
last.plot(kind='density', subplots=True);
```

```python
def distplot(col='price', cut = 1):
    '''
    visualise given column (col) distribution with mean and margin border (cut)
    '''
    hrange = (last[col].max()-last[col].min())/2
    cuts = hrange*cut
    last[col].plot(kind='density')
    plt.plot((last[col].mean(),last[col].mean()),(0, 0.05), c='red')
    plt.plot((last[col].mean()+cuts,last[col].mean()+cuts),(0, 0.05), c='red')
    plt.plot((last[col].mean()-cuts,last[col].mean()-cuts),(0, 0.05), c='red')
    plt.text(last[col].mean(), 0.051, 'Mean', color='r')
    plt.text((last[col].mean()+cuts), 0.051, ('+{}% mean').format(cut), color='r', ha= 'left', va='bottom')
    plt.text(last[col].mean()-cuts, 0.051, ('-{}% mean').format(cut), color='r')
    plt.show()
```

```python
distplot(cut=1.2)
```
