#### From existing db get the url and retrieve the detailed ad's body into a single file - ads.csv

```python
# simple webpage retrival
def simple(url):
    '''
    Returns the page <body> tag from url retrieval
    '''
    page = requests.get(url, headers={'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64; rv:62.0) Gecko/20100101 Firefox/62.0'})
    body = bs(page.content, 'lxml').body

    return body

def details(urls):
    '''
    Go through the list of urls, retrieves the body tag. 
    Sorted by url, saves the body tag of those that has the keyword specific to the url source 
    
    '''
    
    adshu = []
    adscom = []
    adsott = []
    n=0
    for i in urls:
        print(n,'\t',i)
        if 'ingatlanok.hu' in i:
            try:
                back = simple(i)               
                txt = back.get_text()
                
                if 'Azonosító:' in txt:             
                    adshu.append(back)
                
            except: pass
            
        elif 'oc.hu' in i:
            try:
                back = simple(i)               
                txt = back.get_text()
                
                if 'Regisztrációs szám:' in txt:             
                    adsott.append(back)
            except: pass
            
        else:
            try:
                back = simple(i)               
                txt = back.get_text()
                
                if 'Parkolás' in txt or 'Ingatlan állapota' in txt:             
                    adscom.append(back)
            except: pass
        
        n+=1 # counter
    
    now = dtime.datetime.now()
    m = [str(now.month) if len(str(now.month)) == 2 else '0' + str(now.month)]
    d = [str(now.day) if len(str(now.day)) == 2 else '0' + str(now.day)]
    y = str(now.year)
    
    with open('adshu'+ '_' + d[0] + m[0] + y + '.csv', 'w') as f:
                f.write('%s' % adshu)
    with open('adscom'+ '_' + d[0] + m[0] + y + '.csv', 'w') as f:
                f.write('%s' % adscom)
    with open('adsott'+ '_' + d[0] + m[0] + y + '.csv', 'w') as f:
                f.write('%s' % adsott)
    return
```
