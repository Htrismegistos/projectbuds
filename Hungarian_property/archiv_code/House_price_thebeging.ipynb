```python

fname = 'ingatlanCom'
f=open(fname, 'r').read()

soup = BeautifulSoup(f, 'lxml')
csvtext = soup.find_all('div', attrs={'class': re.compile('listing js-listing*')})

data = {}
data['id'] = [i['data-id'] for i in csvtext]
data['price_M_Ft']=[float(i.find('sup').next.next.next.rstrip(' M Ft').replace(" ","")) for i in csvtext]
data['Ft_m2']=[int(i.find(attrs={'class':'listing__price--sqm'}).text.rstrip(' Ft/m2*').replace(" ","")) for i in csvtext]
data['address'] =[i.find('div', attrs={'class':'listing__address'}).text for i in csvtext]
data['district'] =[i.find('div', attrs={'class':'listing__address'}).text.split(' ')[-2] for i in csvtext]
data['area_m2'] = [int(i.find('span',attrs={'class':'listing__data-number'}).text) for i in csvtext]
data['room'] = [i.find(attrs={'class':'listing__save js-press js-save-listing'}).previous\
.previous.previous.previous for i in csvtext]
data['date'] = [dtime.datetime(2018,9,10).strftime("%Y-%m-%d") for i in csvtext]
data['listing_count']= [1 for i in csvtext]

cols = data.keys()
dfCom  = pd.DataFrame(data, columns= cols).sort_values('area_m2').reset_index(drop=True)
```

```python
f = open('ingatlanCom_p1_0922.csv', 'r', encoding= 'UTF-8').read()
csvtext = BeautifulSoup(f)
data = {'id':[],'price M Ft':[],'Ft/m2':[],'address':[], \
    'district': [], 'area m2':[],'room':[],'date':[],'listing_count':[]}
n=1
while n <=14:
    #masodik Com mentes
    f = open('ingatlanCom_p'+ str(n) + '_0922.csv', 'r', encoding= 'UTF-8').read()
    csvtext = BeautifulSoup(f)
    

    
    for i in csvtext:
        a = i.find_all('div', attrs={'class': re.compile('listing js-listing*')})
        #print ('Ads on the page: %d' % len(a))
        data['id']= data['id']+[ap.attrs['data-id'] for ap in i.find_all('div', attrs={'class':re.compile('listing js-listing*')})]
        data['price M Ft']= data['price M Ft']+[float(p.text.lstrip("\n   ").rstrip(' M Ft\n        ')) for p in i.find_all('div', attrs={'class':'price'})]
        data['Ft/m2']=data['Ft/m2']+[int(s.text.lstrip('\n         ').rstrip(' Ft/m\n         \n          2\n         \n').replace(" ","")) for s in i.find_all('div', attrs={'class':'price--sqm'})]
        data['address']=data['address']+[a.text.lstrip('\n        ').rstrip('\n       ') for a in i.find_all('div', attrs={'class':'listing__address'})]
        data['district'] =data['district']+[a.text.split(' ')[-9] for a in i.find_all('div', attrs={'class':'listing__address'})]
        data['area m2']=data['area m2']+[int(a.text.lstrip('\n       ').rstrip(' m² terület\n      ')) for a in i.find_all('div', attrs={'class':'listing__parameter listing__data--area-size'})]
        data['room']=data['room']+[a.text.lstrip('\n       ').rstrip(' szoba\n      ') for a in i.find_all('div', attrs={'class':'listing__parameter listing__data--room-count'})]
        data['date']=data['date']+[dtime.datetime(2018,9,22).strftime("%Y-%m-%d") for a in i.find_all('div', attrs={'class':'price'})]
        data['listing_count']=data['listing_count']+[1 for a in i.find_all('div', attrs={'class':'price'})]

    n+=1

dfCom2 = pd.DataFrame(data, columns= data.keys())
```

```python
f = open('ingatlanHU0910.html', 'r').read()

soup = BeautifulSoup(f)
csvtext = soup.find_all(attrs={'class':'result-row-container'})

data ={}
data['district'] =[ i.find(attrs={'class':'result-title left'}).text.split(' ')[3] for i in csvtext]
data['price'] = [int(i.find(attrs={'class':'result-price left'}).find('a').text.rstrip(' Ft').replace('.','')) for i in csvtext]
data['area m2'] = [int(i.find(attrs={'class':'column-first'}).find('a').next.next.next.text.split(' ')[0]) for i in csvtext]
data['rooms'] = [i.find(attrs={'class':'column'}).find('a').next.next.next.next for i in csvtext]
data['type'] = [i.find(attrs={'class':'result-item-bottom'}).previous.previous.previous.previous.previous.previous.previous for i in csvtext]
data['description'] = [i.find(attrs={'class':'result-short-description'}).text.lstrip("\\n") for i in csvtext]
data['date'] = [dtime.datetime(2018,9,10).strftime("%Y-%m-%d") for i in csvtext]
data['listing_count']= [1 for i in csvtext]

cols = data.keys()
dfHU= pd.DataFrame(data, columns=cols)
df = dfHU.copy()
```
